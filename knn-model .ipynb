```python
col=data.select_dtypes(include='number').columns.tolist()
features=data[col]
features.drop('diagnosis',axis=1,inplace=True)
features
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the features
scaler = StandardScaler()
features = scaler.fit_transform(features)
features
```

```python
target=data['diagnosis']
target
```

```python
# Removing outliers based on IQR for a specific column
Q1 = data['diagnosis'].quantile(0.25)
Q3 = data['diagnosis'].quantile(0.75)
IQR = Q3 - Q1
filtered_data = data[(data['diagnosis'] >= Q1 - 1.5 * IQR) & (data['diagnosis'] <= Q3 + 1.5 * IQR)]
```

```python
filtered_data
```

```python
import sklearn.model_selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.8,random_state=0)
```

```python
x_train
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)

```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, prediction)
print(f'Accuracy of the k-NN model: {accuracy * 100:.2f}%')
```
